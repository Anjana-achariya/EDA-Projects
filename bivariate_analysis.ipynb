```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_df = sns.load_dataset("titanic")
print(titanic_df)
```

```python
# To change the Column names
titanic_df = titanic_df.rename(columns = {
    'pclass':'ticket_class',
    'ticker' : 'ticker_number',
    'cabin' : 'cabin_number',
})
```

```python
# Replace the values
titanic_df["embarked"].replace({
    "C": "Cherbourg",
    "Q": "Queenstown",
    "S" : "Southampton"
}, inplace=True)
```

```python
titanic_df.isnull().sum()
```

```python
# Removing all rows which are having missing values.
titanic_df.dropna()

# Removing only that rows where entire columns have missing values.
titanic_df.dropna(how="all")

# Remove columns with any null values
titanic_cd = titanic_df.dropna(axis=1)
```

```python
sns.kdeplot(x='pclass', hue='fare', data=titanic_df).set(
    title="Relationship between the fare price per gender and per Pclass",
)

plt.show()
```

```python
sns.barplot(x='pclass',y='fare',data= titanic_df)
```

```python
sns.kdeplot(x="pclass", y="fare", data=titanic_df, cmap="Blues", shade=True)

# Add labels and title
plt.xlabel("PClass")
plt.ylabel("Fare")
plt.title("Bivariate KDE Plot")

# Show the plot
plt.show()





```

```python
titanic_hm= titanic_df.corr()

sns.set(rc={'figure.figsize':(8,5)})
sns.heatmap(titanic_hm, annot=True, cmap='rocket')
plt.show()
```
